```python
import zipfile
import tempfile
import matplotlib.pyplot as plt
```

```python
with zipfile.ZipFile('../data/flights/TestFlight_10.29.23.zip') as z:
    for name in z.namelist():
        content = z.read(name)
        with tempfile.NamedTemporaryFile() as tmp:
            tmp.write(content)
            tmp.flush()
            df = decode_blackbox(tmp.name)
            df.to_csv('../data/flights/TestFlight_10.29.23' + name + ".csv", index=False)
```
