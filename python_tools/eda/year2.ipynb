```python
from datetime import datetime, timedelta, timezone
def to_datetime(timestamp: int, epoch = None) -> datetime:
    epoch = epoch or datetime.now(timezone.utc)
    cookie_microseconds_since_epoch = timestamp
    cookie_datetime = epoch + timedelta(microseconds=cookie_microseconds_since_epoch)
    return cookie_datetime
df.apply(lambda row: to_datetime(row['time (us)']), axis=1)
```

```python
run_dfs: dict[str, pd.DataFrame] = dict()
with zipfile.ZipFile(THRUST_ZIP, "r") as zip_ref:
    for run in zip_ref.namelist():
        if run.endswith(".stc") or run.endswith("Notes.txt") or not run.endswith(".txt") or run.startswith("__MACOSX"):
            continue
        print(run)
        text = zip_ref.read(run).decode()
        text = text.split('Torque Measurement System Startup is complete')[1]
        text = "loopIteration,thrust (g),torque,useless_maybe\n" + text
        df = pd.read_csv(StringIO(text))
        df["thrust (g)"] *= -1
        df["torque"] *= -1
        if "/" in run:
            parts = run.split("/")
            run = parts[-1]
        # convert loopIteration to integer
        df["loopIteration"] = df["loopIteration"].astype(int)
        run_dfs[run.split(".")[0].replace('2024', '24').replace('220KV', 'KV220').replace('19X10E', '19X10EX12E').upper()] = df
```

```python
def get_first_peak(series: pd.Series, threshold_factor=0.1) -> int:
    threshold = series.max() * threshold_factor
    smoothing = len(series)//200
    peak = int(series.abs().rolling(smoothing).mean().rolling(2).apply(lambda x: list(x)[1] < list(x)[0] and list(x)[1] > threshold).idxmax()) - 1
    return peak - smoothing
```

```python
for run in runs:
    print(run)
    fig, axs = plt.subplots(1, 2, figsize=(20, 5))

    axs[0].plot(log_dfs[run]['amperage (A)'])
    axs[0].axvline(x=max(get_first_peak(log_dfs[run]['amperage (A)']), get_first_peak(log_dfs[run]['escRPM'])), color='r')

    axs[1].plot(run_dfs[run]['torque'])
    axs[1].axvline(x=get_first_peak(run_dfs[run]['torque']), color='r')
    
    plt.show()
```

```python
merged_dfs: dict[str, pd.DataFrame] = dict()
peak_df = pd.DataFrame(columns=['run', "max_abs_torque", 'min_torque', 'max_torque', 'min_thrust (g)', 'max_thrust (g)', 'max_RPM', 'max_current', 'max_voltage', 'max_throttle'])
for run in runs:
    log_df = log_dfs[run][['loopIteration', 'escRPM', 'amperage (A)', 'vbat (V)', 'rcData[3]']]
    log_df.columns = ['loopIteration', 'escRPM', 'current', 'voltage', 'throttle']
    run_df = run_dfs[run]

    log_sync_feature = max(get_first_peak(log_dfs[run]['amperage (A)']), get_first_peak(log_dfs[run]['escRPM']))
    run_sync_feature = get_first_peak(run_dfs[run]['torque'])

    run_df['synced_iteration'] = run_df['loopIteration'] - run_df.loc[run_sync_feature, 'loopIteration']
    log_df['synced_iteration'] = log_df['loopIteration'] - log_df.loc[log_sync_feature, 'loopIteration']

    merged_dfs[run] = pd.merge_asof(run_df, log_df, on='synced_iteration', direction='nearest')
    peak_df = pd.concat(
        [
            peak_df,
            pd.DataFrame.from_records(
                [
                    {
                        "run": run,
                        "max_abs_torque": run_df['torque'].abs().max(),
                        "min_torque": run_df['torque'].min(),
                        "max_torque": run_df['torque'].max(),
                        'min_thrust (g)': run_df['thrust (g)'].min(),
                        "max_thrust (g)": run_df['thrust (g)'].max(),
                        "max_RPM": log_df['escRPM'].max(),
                        "max_current": log_df['current'].max(),
                        "max_voltage": log_df['voltage'].max(),
                        "max_throttle": log_df['throttle'].max(),
                    }
                ]
            ),
        ],
        ignore_index=True,
    )
```
